```python
import os
import pandas as pd
import numpy as np
from etils import epath
import audio_metadata
```

```python
basepath = epath.Path('/home/jdumbacher/Caples_Creek_project/')
```

```python
recording_names = set()
recording_paths = set()

min_file_size = 100_000_000 # 100 MB or smth like that

for dirpath, dirnames, filenames in os.walk(basepath):
    for filename in filenames:
        if filename.endswith('.wav'):
            audio_path = os.path.join(dirpath, filename)
            filesize = os.path.getsize(audio_path)
            audio_name = filename.split('.')[0] + '.wav' # get rid of .pt things

            if len(filename) < len('26-LEIA_20210529_190000'): # filter out bad names
                continue
            
            if filesize < min_file_size: # check to see if valid ARU recording
                continue
            
            if audio_name in recording_names: # check to see if already added
                continue

            recording_paths.add(audio_path)
            recording_names.add(audio_name)
```
